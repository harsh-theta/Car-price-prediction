## Handling Outliers (Selling Price)

### Outlier Analysis

```python
## Outliers
q1 = df['selling_price'].quantile(0.25)
q3 = df['selling_price'].quantile(0.75)
iqr = q3 - q1
min_price = q1 - (1.5 * iqr)
max_price = q3 + (1.5 * iqr)
```

```python
q1
```

```python
min_outliers = 0
max_outliers = 0

for index, row in df.iterrows():
    if row['selling_price'] < min_price:
        min_outliers +=1
    elif row['selling_price'] > max_price:
        max_outliers += 1

print(f"Outliers more than {max_price}: {max_outliers}")
print(f"Outliers less than {min_price} : {min_outliers}")
```

### Outlier Removal 

```python
removal_indices = []

for index, row in df.iterrows():
    if row['selling_price'] < min_price or row['selling_price'] > max_price:
        removal_indices.append(index)

print(f"Rows before removal: {df.shape}")
df = df.drop(removal_indices)
print(f"Removed {len(removal_indices)} rows")
print(f"New shape: {df.shape}")
```

### Histogram of Selling Price
After removing outliers, we get better distribution of selling price

```python
sns.kdeplot(df["selling_price"], fill=True)
```

## Handling Outliers (Vehicle Age)

### Outlier removal 
As seen in the histogram of vehicle age,
Some cars were older than 15 years, Number of such old cars was very low. Thus, creating skewness in the dataset.
I decided on removing rows with cars older than 15 years.
The decision was taken by visual of the histogram rather than Quantile method.

```python
## Dropping rows with very old cars (age > 15 years)

df = df.drop([index for index,row in df.iterrows() if row['vehicle_age']>15])
```

### Histogram of Vehicle Age

```python
sns.kdeplot(df["vehicle_age"], fill=True)
```

## Handling Outliers (km_driven)

### Outlier analysis

```python
outliers = 0
q1 = df['km_driven'].quantile(0.25)
q3 = df['km_driven'].quantile(0.75)
iqr = q3 - q1
max_limit = q3 + (1.5 * iqr)

for index, row in df.iterrows():
    if row['km_driven'] > max_limit:
        outliers += 1

print(f"Total rows: {df.shape}")
print(f"Outliers with Km_driven feature: {outliers}")
```

### removing the rows with outliers

```python
df = df.drop([index for index, row in df.iterrows() if row['km_driven'] > max_limit])
print(f"Rows dropped.\nNew length is {df.shape}")
```

```python
sns.kdeplot(df["km_driven"], fill=True)
```
